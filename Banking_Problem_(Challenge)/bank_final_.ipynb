## Classifying age in to multiple bins or segments , to simplify a lot of values 

```python

def age(df):
    df.loc[df['age'] <= 32, 'age'] = 1
    df.loc[(df['age'] > 32) & (df['age'] <= 47), 'age'] = 2
    df.loc[(df['age'] > 47) & (df['age'] <= 70), 'age'] = 3
    df.loc[(df['age'] > 70) & (df['age'] <= 98), 'age'] = 4
           
    return df
```

## Similarly classiifying duration into multiple bins to simplify

```python
def duration(df):

    df.loc[df['duration'] <= 102, 'duration'] = 1
    df.loc[(df['duration'] > 102) & (df['duration'] <= 180)  , 'duration']    = 2
    df.loc[(df['duration'] > 180) & (df['duration'] <= 319)  , 'duration']   = 3
    df.loc[(df['duration'] > 319) & (df['duration'] <= 644.5), 'duration'] = 4
    df.loc[df['duration']  > 644.5, 'duration'] = 5

    return df
```

## Using Slices and Voting Classifiers

```python
## slicing with equal parts
df_slice_final1=pd.concat([df_majority.iloc[0:4000],df_minority])
df_slice_final2=pd.concat([df_majority.iloc[4000:8000],df_minority])
df_slice_final3=pd.concat([df_majority.iloc[8000:12000],df_minority])
df_slice_final4=pd.concat([df_majority.iloc[12000:16000],df_minority])
df_slice_final5=pd.concat([df_majority.iloc[16000:20000],df_minority])
df_slice_final6=pd.concat([df_majority.iloc[20000:24000],df_minority])
df_slice_final7=pd.concat([df_majority.iloc[24000:28000],df_minority])
df_slice_final8=pd.concat([df_majority.iloc[28000:32000],df_minority])
df_slice_final9=pd.concat([df_majority.iloc[32000:36548],df_minority])


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test = train_test_split(df_slice_final2.drop(['y'], axis = 1), df_slice_final2["y"], test_size = 0.3, random_state = 42)
clf1 = LogisticRegression(solver='lbfgs', multi_class='multinomial',random_state=42)
clf2 = RandomForestClassifier(n_estimators=200, random_state=123,class_weight="balanced")
clf3 = KNeighborsClassifier()

eclf1 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('knn', clf3)], voting='hard')
eclf1 = eclf1.fit(X_train, y_train)
y_pred1= eclf1.predict(X_test)

print(classification_report(y_test, y_pred1))
print(accuracy_score(y_test, y_pred1))
f1_score(pd.get_dummies(y_test,drop_first=True),pd.get_dummies(y_pred1,drop_first=True))

eclf2 = VotingClassifier(estimators=[ ('lr', clf1), ('rf', clf2), ('knn', clf3)], voting='soft')
eclf2 = eclf2.fit(X_train, y_train)
y_pred2=eclf2.predict(X_test)

print(classification_report(y_test, y_pred2))
print(accuracy_score(y_test, y_pred2))
f1_score(pd.get_dummies(y_test,drop_first=True),pd.get_dummies(y_pred2,drop_first=True))

eclf3 = VotingClassifier(estimators=[ ('lr', clf1), ('rf', clf2), ('knn', clf3)], voting='soft', weights=[2,1,1], flatten_transform=True)
eclf3 = eclf3.fit(X_train, y_train)
y_pred3=eclf3.predict(X_test)

print(classification_report(y_test, y_pred3))
print(accuracy_score(y_test, y_pred3))
f1_score(pd.get_dummies(y_test,drop_first=True),pd.get_dummies(y_pred3,drop_first=True))
```
